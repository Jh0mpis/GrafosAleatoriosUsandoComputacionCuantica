# Circuito usado para la encriptación cuántica 

Para el caso de encriptación de una palabra se pasó cada letra de la palabra a un valor en binario, debido a que los qbits son la extensión cuántica de los bits clásicos, es posible escribir un número binario con qbits usando la compuerta $X$. 

Por ejemplo, el siguiente circuito es la representación del número "0" enn binario

```python
circuit = QuantumCircuit(1,1)
circuit.draw(output = "mpl")
```

mientras que el número "1" vendrá dado como

```python
circuit.x(0)
circuit.draw(output = "mpl")
```

Para el caso de una palabra se tiene el siguiente ejemplo
a$\rightarrow 00000$,
c$\rightarrow 00010$,
e$\rightarrow 00101$,
para cada letra hay 5 qbits que la representan, por ejemplo, para el caso de la e, el circuito será como sigue

```python
circuit = QuantumCircuit(5,5)
circuit.x(2)
circuit.x(4)
circuit.draw(output = "mpl")
```

# Compuerta de Hadamard

Esta compuerta cuántica permite superponer un qbit en los estados 1 y 0 con igual probabilidad, esto es muy útil para generar procesos aleatorios, esta compuerta será usada para realizar una transformación sobre la palabra de entrada. 

El circuito para un valor binario arbitrariio viene dado como sigue 

```python
prepareCircuit("01101").draw(output ="mpl")
```

Este será el circuito sobre el cual se medirá para obtener un valor aleatorio de cada letra ingresada, la medida sobre varios circuitos determinará el cambio de toda la palabra.
